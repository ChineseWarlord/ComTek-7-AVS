```python
file = open('data/test_dataset.txt', 'r')
test_data = file.read().split('\n')
del test_data[-1]

for i, image in enumerate(test_data):
    annotation = image.replace('clips', 'annotations').replace('frame', 'annotations').replace('.jpg', '.txt')
    test_data[i] = annotation
```

```python
humans = 0
bicycles = 0
motorcycles = 0
vehicles = 0

for file in test_data:
    with open(file, 'r') as f:
        anno_file = f.read()
        humans += anno_file.count('human')
        bicycles += anno_file.count('bicycle')
        motorcycles += anno_file.count('motorcycle')
        vehicles += anno_file.count('vehicle')
```

```python
print('Humans:', humans, '\nBicycles:', bicycles, '\nMotorcycles:', motorcycles, '\nVehicles', vehicles)
```
