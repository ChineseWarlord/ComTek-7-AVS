Background subtraction:

```python
# Load imgs
images = glob.glob(data + "*.tif")
 
# BG image
bg = cv.imread(images[0], cv.IMREAD_GRAYSCALE)

# Set the desired window size
window_width = 800  # Width in pixels
window_height = 600  # Height in pixels

# Create a window with the specified size
cv.namedWindow('Background Subtraction', cv.WINDOW_NORMAL)
cv.resizeWindow('Background Subtraction', window_width, window_height)

# Last frame flag
last_frame = False 

# Flip through imgs
for image in images:
    img = cv.imread(image, cv.IMREAD_GRAYSCALE)
    
    # Find the abs difference
    abs_diff = cv.absdiff(img,bg)
    
    # Apply adaptive or simple thresholding
    #threshold = cv.adaptiveThreshold(abs_diff,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,3,2)
    #_, threshold = cv.threshold(abs_diff, 10, 255, cv.THRESH_BINARY)
    
    cv.imshow("Background Subtraction",abs_diff)
    #cv.imshow("Window",threshold)
    
    k = cv.waitKey(0)
    if k == ord("q"):
        break
    
    # If reached last element flip last_frame
    if image == images[-1]:
        last_frame = True

# Keep displaying the last frame
if last_frame == True:
    while 1:
        k = cv.waitKey(0)
        if k == ord("q"):
            break
    
cv.destroyAllWindows()   
print("Successfully exited!")
```

Image differencing:

```python
# Load imgs
images = glob.glob(data + "*.tif")

# Set the desired window size
window_width = 800  # Width in pixels
window_height = 600  # Height in pixels

# Create a window with the specified size
cv.namedWindow('Image Differencing', cv.WINDOW_NORMAL)
cv.resizeWindow('Image Differencing', window_width, window_height)

# Last frame flag
last_frame = False

# Define previous frame
prev_frame = None

# Flip through imgs
for image in images:
    img = cv.imread(image, cv.IMREAD_GRAYSCALE)
    
    if prev_frame is not None:
        # Find the abs difference
        abs_diff = cv.absdiff(img,prev_frame)
        
        # Apply adaptive or simple thresholding
        #threshold = cv.adaptiveThreshold(abs_diff,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,3,2)
        _, threshold = cv.threshold(abs_diff, 10, 255, cv.THRESH_BINARY)
        
        cv.imshow("Image Differencing",abs_diff)
        #cv.imshow("Window",threshold)
    else:
        cv.imshow("Image Differencing",img)
    
    k = cv.waitKey(0)
    if k == ord("q"):
        break
    
    # If reached last element flip last_frame
    if image == images[-1]:
        last_frame = True
        
    prev_frame = img

# Keep displaying the last frame
if last_frame == True:
    while 1:
        k = cv.waitKey(0)
        if k == ord("q"):
            break
    
cv.destroyAllWindows()   
print("Successfully exited!")
```

Image differencing + highlighting:

```python
# Load imgs
images = glob.glob(data + "*.tif")

# Set the desired window size
window_width = 800  # Width in pixels
window_height = 600  # Height in pixels

# Create a window with the specified size
cv.namedWindow('Image Differencing + Highlight', cv.WINDOW_NORMAL)
cv.resizeWindow('Image Differencing + Highlight', window_width, window_height)

# Flags
last_frame = False
prev_frame = None
adaptive = True

# Threshold, BlockSize, Constant C
th_value = 1
C = 0
blockSize = 3 # Must be 3x3,7x7,9x9 ...

# Flip through imgs
for image in images:
    img = cv.imread(image, cv.IMREAD_GRAYSCALE)
    
    if prev_frame is not None:
        # Find the abs difference
        abs_diff = cv.absdiff(img,prev_frame)
        
        
        # Apply adaptive or simple thresholding
        if adaptive == True:
            threshold = cv.adaptiveThreshold(abs_diff,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,blockSize,C)
        else:
            _, threshold = cv.threshold(abs_diff, th_value, 255, cv.THRESH_BINARY)
        
        # Create a 3-channel image with red highlighting where differences are found
        highlighted_diff = cv.cvtColor(img, cv.COLOR_GRAY2BGR)
        highlighted_diff[threshold > 0] = [0, 0, 255]  # Set the pixels to red
        
        cv.imshow("Image Differencing + Highlight",highlighted_diff)
    else:
        cv.imshow("Image Differencing + Highlight",img)
    
    k = cv.waitKey(0)
    if k == ord("q"):
        break
    
    # If reached last element flip last_frame
    if image == images[-1]:
        last_frame = True
        
    prev_frame = img

# Keep displaying the last frame
if last_frame == True:
    while 1:
        k = cv.waitKey(0)
        if k == ord("q"):
            break
    
cv.destroyAllWindows()   
print("Successfully exited!")
```

Background subtraction with subtractor object:

```python
# Load imgs
images = glob.glob(data + "*.tif")
 
# BG image
bg = cv.imread(images[0], cv.IMREAD_GRAYSCALE)

# Set the desired window size
window_width = 800  # Width in pixels
window_height = 600  # Height in pixels

# Create a window with the specified size
cv.namedWindow('Background Subtraction with subtractor object', cv.WINDOW_NORMAL)
cv.resizeWindow('Background Subtraction with subtractor object', window_width, window_height)

# Last frame flag
last_frame = False 

# Initialize bg subtractor object
backSub = cv.createBackgroundSubtractorMOG2()

# Flip through imgs
for image in images:
    img = cv.imread(image)
    
    # Apply bg subtractor
    img = backSub.apply(img)
    
    ## Apply threshold
    #_, threshold = cv.threshold(img, 0, 255, cv.THRESH_BINARY)
    
    ## Create a 3-channel image with red highlighting where differences are found
    #img = cv.cvtColor(img, cv.COLOR_GRAY2BGR)
    #img[threshold > 0] = [0, 0, 255]  # Set the pixels to red
    
    cv.imshow("Background Subtraction with subtractor object",img)
    
    k = cv.waitKey(0)
    if k == ord("q"):
        break
    
    # If reached last element flip last_frame
    if image == images[-1]:
        last_frame = True

# Keep displaying the last frame
if last_frame == True:
    while 1:
        k = cv.waitKey(0)
        if k == ord("q"):
            break
    
cv.destroyAllWindows()   
print("Successfully exited!")
```
